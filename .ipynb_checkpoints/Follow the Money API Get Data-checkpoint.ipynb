```python
import pandas as pd
import numpy as np

import json
from urllib.request import urlopen
from pandas.io.json import json_normalize 
import pickle


```

```python
# function that returns a json
def return_json(link):
    jsonurl = urlopen(link)
    js = json.load(jsonurl)
    return js


# json to pandas
def parse_json_page(json):
    
    # get number of rows
    rows = json['metaInfo']['paging']['recordsThisPage']

    key_columns = list(json['records'][0].keys())
    names_columns = []
    flat_list = []

    # get the column names
    for i in key_columns:
        try: 
            keys = json['records'][0][i].keys()
            
            if len(keys) == 1:
                names_columns.append(keys[0])
            else:
                for j in keys:
                    names_columns.append(str(i) + "_" + str(j))
        except:
            names_columns.append(i)

    # get the value from the rows
    for i in range(0, rows):
        for j in key_columns:
            try: 
                keys = json['records'][i][j].keys()
                for k in keys:
                    flat_list.append(json['records'][i][j][k])
            except:
                flat_list.append(json['records'][i][j])

    two_array = np.array(flat_list)   
    
    # make and return the dataframe
    return pd.DataFrame(two_array.reshape(rows, len(names_columns)), columns = names_columns)


# gets all of the values for the given api call to pandas
# part1 is the first part of the link
# part2 is the second part of the link
def parse_all(part1, part2):

    # find the number of pages
    link = part1 + '&p=' + str(0) + part2 
    j = return_json(link)

    max_page = j['metaInfo']['paging']['maxPage']
    dict_page = {}
    print(max_page)

    # iterate through each page making a df for each page in a dict
    for i in range(0, max_page + 1):
        print(i)
        new_link = part1 + '&p=' + str(i) + part2
        json_page = return_json(new_link)
        df_page = parse_json_page(json_page)
        dict_page[i] = df_page

    # merge all the dataframe values together and return 
    return pd.concat(dict_page.values(), ignore_index = True)


```
