```python
def dtype_consistent(df):
    column_vals = df.columns.tolist()
    for i in column_vals:
        if i not in dtypes_df:
            print('what column names are you using?')
            continue
        if df[i].dtypes != dtypes_df[i]:
            df[i] = df[i].astype(dtypes_df[i])
    return df
            
    


# function that takes a list of seperated csv and combines them after checking that datatypes are the same
# filenames must be a list
# name must be string

# this will mostly be for contributions by indididuals since it is split
def load_sep_files(path, filenames, name, header_name):
    # make a dict
    container = {}
    counter = 0
    header = pd.read_csv(header_name, header = None)
    for filename in filenames:
        container[name + ascii_lowercase[counter]] = pd.read_csv(path + filename, sep= "|", header = None)
        counter = counter + 1
    
    # give each df column names
    for item in container:
        container[item].columns = header.values[0].tolist()
        # check to make sure datatypes are right
        item = dtype_consistent(item)
        
    
    # merge data_frames
    df_merged = container[name + ascii_lowercase[0]]

    for  i in range(1, counter):
        df_merged = df_merged.append(container[name + ascii_lowercase[i]])
        
    return df_merged
```
